```python
import re

names = {}

for idx, row in data.iterrows():
    match = re.findall(r'.+,\s*(?:(?:Mrs)|(?:Miss))\.[\s]*(?:(\w*))(?:[\s\w]*\((\w+))?', row['Name'])
    if match:
        # print(match)
        name = match[0][1] if match[0][1] else match[0][0]
    else:
        continue
        
    try:
        names[name] += 1
    except KeyError:
        names[name] = 1
  
    
print('\n'.join(f'{k}: {v}' for k, v in sorted(names.items(), key=lambda x: x[1], reverse=True)))
```

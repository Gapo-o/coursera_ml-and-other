```python
from copy import deepcopy
import pandas
import numpy as np

import sklearn
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor


x = sklearn.datasets.load_boston() 

print(list(x))

# print(x)

data = deepcopy(x["data"])
target = deepcopy(x["target"])

print (data)
print (target)
```

```python
st_x = sklearn.preprocessing.scale(data)
st_y = sklearn.preprocessing.scale(target)
print(st_x)
```

```python
p_values = np.linspace(1, 10, num= 200)

print(p_values)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

reg = KNeighborsRegressor(n_neighbors=5, weights="distance")

result = cross_val_score(reg, st_x, st_y, cv = kf, scoring = 'neg_mean_squared_error')

print(result)

print(np.mean(result))

result_max = p_values[int(max(result))]

print (result_max)

```

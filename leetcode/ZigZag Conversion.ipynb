```python
numRows = 4
s = "abcde"


tmp = len(s)
col = 0

if numRows == 1:
    print(s)

if tmp <= numRows:
    col = tmp
else:
    while (1):
        tmp = tmp - (2 * numRows - 2)
        col = col + numRows - 1
        if tmp < 0:
            break
        elif 0 <= tmp <= numRows:
            col = col + 1
            break

arr = [["" for j in range(col)] for i in range(numRows)]

if numRows == 2:
    j = 0
    for i in range(col):
        for g in range(numRows):
            if j >= len(s):
                break
            arr[g][i] = s[j]
            j = j + 1
else:
    j = 0
    flag = 0
    row_ = numRows - 1
    for i in range(col):
        if flag == 0:
            for g in range(numRows):
                arr[g][i] = s[j]
                j = j + 1
                if j >= len(s):
                    break

            flag = flag + 1
        else:
            arr[row_ - 1][i] = s[j]
            j = j + 1
            if j >= len(s):
                break
            row_ = row_ - 1
            flag = flag + 1

            if flag == numRows - 1:
                row_ = numRows - 1
                flag = 0

j = 0
answer = [""] * len(s)
for i in range(numRows):
    for g in range(col):
        if arr[i][g] == "":
            continue
        else:
            answer[j] = arr[i][g]
            j += 1

import pandas
df = pandas.DataFrame(arr)
print(df)
print(answer)


# class Solution:
#     def convert(self, s: str, numRows: int) -> str:
#
#         tmp = len(s)
#         col = 0
#
#         if numRows == 1:
#             return s
#
#         if tmp <= numRows:
#             col = tmp
#         else:
#             while (1):
#                 tmp = tmp - (2 * numRows - 2)
#                 col = col + numRows - 1
#                 if tmp < 0:
#                     break
#                 elif 0 <= tmp <= numRows:
#                     col = col + 1
#                     break
#
#         arr = [["" for j in range(col)] for i in range(numRows)]
#
#         if numRows == 2:
#             j = 0
#             for i in range(col):
#                 for g in range(numRows):
#                     if j >= len(s):
#                         break
#                     arr[g][i] = s[j]
#                     j = j + 1
#         else:
#             j = 0
#             flag = 0
#             row_ = numRows - 1
#             for i in range(col):
#                 if flag == 0:
#                     for g in range(numRows):
#                         if j >= len(s):
#                             break
#                         arr[g][i] = s[j]
#                         j = j + 1
#
#                     flag = flag + 1
#                 else:
#                     if j >= len(s):
#                         break
#                     arr[row_ - 1][i] = s[j]
#                     j = j + 1
#                     row_ = row_ - 1
#                     flag = flag + 1
#
#                     if flag == numRows - 1:
#                         row_ = numRows - 1
#                         flag = 0
#
#         j = 0
#         answer = [""] * len(s)
#         for i in range(numRows):
#             for g in range(col):
#                 if arr[i][g] == "":
#                     continue
#                 else:
#                     answer[j] = arr[i][g]
#                     j += 1
#
#         return ''.join(answer)

```
